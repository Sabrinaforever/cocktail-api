```python
import pandas as pd

df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Wages.csv", encoding='utf-16', sep='\t')
```

```python
df_cityandwage = df
selected_columns = df_cityandwage[["City (Average workplace wages)", "Wages"]]
print(selected_columns)
```

```python
cleaned_df = selected_columns.dropna(subset=['Wages'])
print(cleaned_df)
```

```python
cities = cleaned_df['City (Average workplace wages)']
wages = cleaned_df['Wages']
plt.figure(figsize=(10, 8))
plt.bar(cities, wages, color=['orange' if city == 'Bristol' else 'skyblue' for city in cities])

plt.xlabel('City')
plt.ylabel('Average Wages')
plt.title('Average Wages by City in the U.K in 2023')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("/content/gdrive/MyDrive/figAverageWagesbyCity.pdf")
plt.show
```

```python
cleaned_df.rename(columns={'City (Average workplace wages)': 'City'}, inplace=True)
cleaned_df2.rename(columns={'City (House prices)': 'City'}, inplace=True)
merged_df= pd.merge(cleaned_df, cleaned_df2, on='City')
print(merged_df)
```

```python
merged_df['House prices £'] = pd.to_numeric(merged_df['House prices £'].astype(str).str.replace(',', ''), errors='coerce')
```

```python
plt.figure(figsize=(10, 20))

indices = np.arange(len(merged_df['City']))
bar_width = 0.35

for i, row in merged_df.iterrows():
    color_wages = 'darkblue' if row['City'] == 'Bristol' else 'skyblue'
    color_prices = 'red' if row['City'] == 'Bristol' else 'orange'
    plt.barh(i - bar_width/2, row['Wages'], height=bar_width, color=color_wages)
    plt.barh(i + bar_width/2, row['House prices £']/1000, height=bar_width, color=color_prices)

plt.yticks(indices, merged_df['City'])
plt.xlabel('Values')
plt.ylabel('Cities')
plt.title('Comparison of Average Wages and House Prices by City in the U.K. 2023')
plt.legend(['Average Wages (£)', 'Average House Prices (£1000s)'], loc='best')

plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("/content/gdrive/MyDrive/combinewageandprice.pdf")
plt.show()
```
